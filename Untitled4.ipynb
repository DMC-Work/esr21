```python
import os, sys
import django
import pytz
from dateutil import parser
from django.utils.timezone import make_aware
os.environ["DJANGO_ALLOW_ASYNC_UNSAFE"] = "true"
from edc_base.utils import age, get_utcnow
sys.path.append('../..') # add path to project root dir
os.environ["DJANGO_SETTINGS_MODULE"] = "esr21.settings"
import pandas as pd

# Connect to Django ORM
django.setup()
import uuid
from django.contrib.sites.models import Site
from django.core.exceptions import ValidationError
```

```python
ps_qs = PregnancyStatus.objects.all().values_list('id','subject_visit__subject_identifier',
    'subject_visit__visit_code', 'report_datetime','start_date_menstrual_period','expected_delivery',
    'contraceptive_usage', 'contraceptive', 'contraceptive_other', 'surgically_sterilized', 'amenorrhea_history',
    'post_menopausal_range', 'post_menopausal','post_menopausal_other', 'child_bearing_potential', 'comment',
    'number_prev_pregnancies','number_normal_pregnancies', 'number_miscarriages', 'date_miscarriages',
    'risk_factor', 'created', 'user_created', 'modified', 'user_modified', 'site_id')
```

```python
final_data = [['id', 'subject_visit__subject_identifier',
    'subject_visit__visit_code', 'report_datetime','start_date_menstrual_period','expected_delivery',
    'contraceptive_usage', 'contraceptive', 'contraceptive_other', 'surgically_sterilized', 'amenorrhea_history',
    'post_menopausal_range', 'post_menopausal','post_menopausal_other', 'child_bearing_potential', 'comment',
    'number_prev_pregnancies','number_normal_pregnancies', 'number_miscarriages', 'date_miscarriages',
    'risk_factor', 'created', 'user_created', 'modified', 'user_modified', 'site_id', 'age', 'gender','contraceptives'],]
for a in ps_qs:
    a = list(a)
    try:
        ps = PregnancyStatus.objects.get(id=a[0])
    except:
        print('hmmm')
    else:
        contraceptives_list = ps.contraceptive.all().values_list('name')
        headings = final_data[0]
        a[headings.index('contraceptive')] = contraceptives_list
        
    try:
        ic = InformedConsent.objects.get(subject_identifier=a[1])
    except InformedConsent.DoesNotExist:
        pass
    else:
        part_age = age(ic.dob, get_utcnow()) 
        if part_age:
            part_age=part_age.years
            
            a.append(part_age)
            final_data.append(a)
        a.append(ic.gender)
```

```python
df_vc = pd.DataFrame(final_data)
df_vc
```
