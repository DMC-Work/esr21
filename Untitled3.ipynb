```python
import os, sys
import django
import pytz
from dateutil import parser
from django.utils.timezone import make_aware
os.environ["DJANGO_ALLOW_ASYNC_UNSAFE"] = "true"
from edc_base.utils import age, get_utcnow
sys.path.append('../..') # add path to project root dir
os.environ["DJANGO_SETTINGS_MODULE"] = "esr21.settings"
import pandas as pd

# Connect to Django ORM
django.setup()
import uuid
from django.contrib.sites.models import Site
from django.core.exceptions import ValidationError
```

```python
cvs = VaccinationDetails.objects.values_list('subject_visit__subject_identifier', 'adverse_event', 'site_id',
                                             'vaccination_site', 'received_dose_before')
```

```python
from edc_base.utils import age, get_utcnow
final_data = [['pid', 'adverse_event', 'site_id', 'vaccination_site', 'dose_received', 'age', 'gender']]
for a in cvs:
    try:
        ic = InformedConsent.objects.get(subject_identifier=a[0])
    except InformedConsent.DoesNotExist:
        pass
    else:
        part_age = age(ic.dob, get_utcnow()) 
        if part_age:
            part_age=part_age.years
            a = list(a)
            a.append(part_age)
            final_data.append(a)
        a.append(ic.gender)
        
```

```python
df_vc = pd.DataFrame(final_data)
df_vc
```
